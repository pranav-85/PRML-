```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris_data = pd.read_csv('iris.csv')
iris_data
```

```python
data = iris_data[['sepal.length', 'sepal.width']].to_numpy()
data
```

```python
def euclidean_distance(x, y):
    return ((x[1] - y[1])**2 + (x[0] - y[0])**2)**0.5

pairwise_distances = []

for pt1 in data:
    pt_dist = []
    for pt2 in data:
        pt_dist.append(euclidean_distance(pt1, pt2))
    pairwise_distances.append(pt_dist)

pairwise_distances
```

```python
def prim_MST(distances):
    n = len(distances)
    point = 0  # Start from the first node
    visited = [False] * n  # Keep track of visited nodes
    visited[point] = True  # Mark the starting node as visited
    edges = []  # To store the edges of the Minimum Spanning Tree (MST)
    
    while len(edges) < n - 1:
        min_edge = None
        min_weight = float('inf')
        
        # Find the minimum weight edge connecting a visited node to an unvisited node
        for i in range(n):
            if visited[i]:  # Check only visited nodes
                for j in range(n):
                    if not visited[j] and distances[i][j] < min_weight:
                        min_weight = distances[i][j]
                        min_edge = (i, j)
        
        if min_edge:
            edges.append(min_edge)
            visited[min_edge[1]] = True  # Mark the new node as visited
    
    return edges
```

```python
edges = prim_MST(pairwise_distances)
edges
```

```python
def get_distance(pairwise_distances, indices):
    return pairwise_distances[indices[0]][indices[1]]

edge_distances = [get_distance(pairwise_distances, edge) for edge in edges]

n = len(edge_distances)
# max_distance_idx = np.argmax(edge_distances)

# edges.pop(max_distance_idx)

# max_distance_idx = np.argmax(edge_distances)
# edges.pop(max_distance_idx)

max_dist_idx = np.flip(np.argsort(edge_distances))[0:2]
max_dist_edges = [edges[i] for i in max_dist_idx]

for edge in max_dist_edges:
    edges.remove(edge)
edges
# edges
```

```python
def visualize_mst_with_clusters(data, edges, clusters, class_names):
    # Create a color map for the clusters
    unique_clusters = np.unique(clusters)
    colors = plt.cm.rainbow(np.linspace(0, 1, len(unique_clusters)))
    cluster_colors = {cluster: color for cluster, color in zip(unique_clusters, colors)}

    # Plot nodes with class colors
    for cluster in unique_clusters:
        indices = np.where(clusters == cluster)
        plt.scatter(data[indices, 0], data[indices, 1], color=cluster_colors[cluster], s=100, 
                    label=class_names[cluster])

    # Plot edges from the MST
    for edge in edges:
        p1, p2 = edge
        x_values = [data[p1, 0], data[p2, 0]]
        y_values = [data[p1, 1], data[p2, 1]]
        plt.plot(x_values, y_values, color='black', zorder=1)

    # Add labels, legend, and show plot
    plt.title('Minimum Spanning Tree with Identified Clusters (Iris Dataset)')
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.legend(loc='best')  # Add legend for class names
    plt.show()

# Example usage:
# Define class names based on Iris dataset's classes
class_names = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}

# Dummy clusters for now (this should be derived from your clustering logic)
clusters = np.zeros(len(data))
clusters[50:100] = 1  # Assign clusters for two classes
clusters[100:] = 2

# Visualize the MST and clusters with class names
visualize_mst_with_clusters(data, edges, clusters, class_names)
```
