```python
import numpy as np
import pandas as pd
```

```python
iris_data = pd.read_csv('iris.csv')

iris_data.isnull().sum()
```

```python
#b)
n = len(iris_data)
columns = iris_data.drop(['variety'], axis=1).columns

r = []
for col1 in columns:
    l = []
    for col2 in columns:
        
        if col1 == col2:
            l.append(1)
        else:
            x_mean = np.mean(iris_data[col1])
            y_mean = np.mean(iris_data[col2])
            
            x_diff = iris_data[col1] - x_mean
            y_diff = iris_data[col2] - y_mean
            numerator = sum(x_diff * y_diff)
            denominator = (sum(x_diff**2) * sum(y_diff**2)) ** 0.5

            val = numerator / denominator
            l.append(val)
    r.append(l)

r
```

```python
import seaborn as sns

sns.heatmap(data=r, annot=True, cmap="coolwarm", xticklabels=columns, yticklabels=columns)
```

### Inferences

- **The correlation between `petal.length` and `petal.width` is the best one for predicting.**
  
- **The correlation between `sepal.width` and `petal.length` is the worst one.**

- In Q2, the accuracy of the model with `sepal.length` and `sepal.width` was only 0.77 for 3 neighbours. Thus, their correlation of -0.12 makes sense.
